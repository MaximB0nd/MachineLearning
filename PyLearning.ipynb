```python
arr1 = [1, 2, 11, 3, 4, 5, 6, 7, 8, 9, 10, 12]
arr2 = [2, 5, 6, 1, 4, 9, 8, 7, 10, 3]
def sort_key(item):
    if item in arr2:
        return  arr2.index(item)
    else:
        return len(arr2)
arr1.sort(key=sort_key)
arr1
```
